```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
df = pd.DataFrame(pd.read_csv(os.path.join(r"C:\Users\rahil\Desktop\Kaggle\Accidents0515.csv")))
df.head()

```

```python
x=df['Local_Authority_(Highway)']
y=df['Number_of_Casualties']
```

```python
plt.figure(figsize=(18,6))
plt.scatter(x[:5500], y[:5500])
plt.title('No. of Casualty per Local Highway Authority ')
plt.xlabel('Local Highway Authority')
plt.ylabel('No.of causualties')
plt.show()
```

```python
y2=df['Light_Conditions']
plt.figure(figsize=(18,6))
plt.scatter(x[:5500], y2[:5500])
plt.title('Light Conditions causing Accidents for Local-Highway Authority ')
plt.xlabel('Local Highway Authority')
plt.ylabel('Light Conditions')
plt.show()
```
